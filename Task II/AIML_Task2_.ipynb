# AI & ML Task 2
## Feature Engineering, Model Optimization & Performance Comparison

### Project Overview
Compare Linear Regression, Ridge Regression, and Decision Tree Regressor on the California Housing Dataset.

## 1. Import Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
import joblib
```

## 2. Load Dataset

```
data = fetch_california_housing(as_frame=True)
df = pd.concat([data.data, data.target.rename('HousePrice')], axis=1)
df.head()
```

## 3. Dataset Information

```
print(df.shape)
df.info()
df.describe()
```

## 4. Missing Value Analysis

```
df.isnull().sum()
```

## 5. Correlation Heatmap

```
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## 6. Population vs House Price

```
plt.figure(figsize=(8,6))
plt.scatter(df['Population'], df['HousePrice'], alpha=0.4)
plt.xlabel('Population')
plt.ylabel('House Price')
plt.title('Population vs House Price')
plt.show()
```

## 7. Median Income vs House Price

```
plt.figure(figsize=(8,6))
plt.scatter(df['MedInc'], df['HousePrice'], alpha=0.4)
plt.xlabel('Median Income')
plt.ylabel('House Price')
plt.title('Median Income vs House Price')
plt.show()
```

## 8. Feature Engineering & Scaling

```
X = df.drop('HousePrice', axis=1)
y = df['HousePrice']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## 9. Train-Test Split

```
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

## 10. Train Multiple Models

```
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(alpha=1.0),
    'Decision Tree': DecisionTreeRegressor(max_depth=5, random_state=42)
}
```

## 11. Model Evaluation and Comparison

```
results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    rmse = np.sqrt(mean_squared_error(y_test, pred))
    r2 = r2_score(y_test, pred)

    results.append([name, rmse, r2])

results_df = pd.DataFrame(results, columns=['Model','RMSE','R2 Score'])
results_df
```

## 12. Model Comparison Charts

```
plt.figure(figsize=(8,5))
sns.barplot(data=results_df, x='Model', y='R2 Score')
plt.title('R2 Score Comparison')
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(data=results_df, x='Model', y='RMSE')
plt.title('RMSE Comparison')
plt.show()
```

## 13. Best Model Visualization

```
best_model = LinearRegression()
best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)

plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.4)
plt.plot([min(y_test), max(y_test)],
         [min(y_test), max(y_test)],
         color='red')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted House Prices')
plt.show()
```

## 14. Residual Analysis

```
residuals = y_test - y_pred

plt.figure(figsize=(8,6))
sns.histplot(residuals, kde=True, bins=30)
plt.title('Residual Distribution')
plt.show()

plt.figure(figsize=(8,6))
plt.scatter(y_pred, residuals, alpha=0.4)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted Values')
plt.show()
```

## 15. Save Best Model

```
joblib.dump(best_model, 'best_model.joblib')
```

## 16. Conclusion
Compared multiple regression models, evaluated them using RMSE and R², performed feature scaling, and selected the best model based on performance.
